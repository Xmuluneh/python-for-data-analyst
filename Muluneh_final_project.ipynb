```python
#Establishing the connection
conn = psycopg2.connect(
   database="final_mul", user='postgres', password='passwd123', host='127.0.0.1', port= '5432')
#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Doping final_mul table if already exists.
cursor.execute("DROP TABLE IF EXISTS final_mul")

#Creating table as per requirement
sql ='''CREATE TABLE final_mul(
   date date,
   location varchar(50),
   mintempâ„ƒ varchar,
   maxtempâ„ƒ varchar,
   rainfall float,
   sunshine varchar,
   raintoday varchar,
   year int,
   month int,
   day int

)'''
cursor.execute(sql)
print("Table created successfully........")
conn.commit()
#Closing the connection
conn.close()
```

```python
# define a function to execute
def execute_values(conn, df, table):
  
    tuples = [tuple(x) for x in df.to_numpy()]
  
    cols = ','.join(list(df.columns))
    # SQL query to execute
    query = "INSERT INTO %s(%s) VALUES %%s" % (table, cols)
    cursor = conn.cursor()
    try:
        extras.execute_values(cursor, query, tuples)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    print("the dataframe is inserted")
    cursor.close()
  
  
conn = psycopg2.connect(
    database="final_mul", user='postgres', password='passwd123', host='127.0.0.1', port='5432'
)
  
df = new_df
  
execute_values(conn, df, 'final_mul')
```
