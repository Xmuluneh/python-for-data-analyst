```python
# 2. Select only `Date`,`Location`, `MinTemp`, `MaxTemp`, `Rainfall`, `Sunshine`, `RainToday` fields that are after `2010`.
Aus_weather = weather_df[['Date','Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Sunshine', 'RainToday']]
Aus_weather= Aus_weather[Aus_weather.Date >= '2010-01-01']
Aus_weather.head()
```

```python
# 3. Rename all the columns to something easier to read and understand.
Aus_weather.rename({'Date': 'Forcast_Date', 'Location':'City', 'MinTemp':'Temp_Min',
'MaxTemp':'Temp_Max', 'Rainfall':'Measured_Rainfall','RainToday':'Is_Rainy_Today'}, axis='columns',  inplace = True)
Aus_weather.head()
```

```python
# 4. Then change the `RainToday=Is_Rainy_Today` values (of cases after you renamed the column) to `0` when is `No` and `1` when it is `Yes`.
Aus_weather['Is_Rainy_Today'] = Aus_weather['Is_Rainy_Today'].replace(['No','Yes'],['0','1'])
Aus_weather.head()
```

```python
# 5. Add a new column `day`, `month`, and `year` column (three new columns) based on `Date` column.
Aus_weather['Day'] = pd.to_datetime(Aus_weather['Forcast_Date']).dt.day
Aus_weather['Month'] = pd.to_datetime(Aus_weather['Forcast_Date']).dt.month
Aus_weather['Year'] = pd.to_datetime(Aus_weather['Forcast_Date']).dt.year
Aus_weather.head()
# try with slice or split
```

```python
 # 7. calculate the `yearlly_average` minimum and maximum temperature
 # 8. then fill the missing values by what you calculate as the yearly average minimum and maximum temperature.
  #    example: 
  #      - say the average `2020` temperature was `16^c`. so this means any 2020 record that missed a value would have been replaced by the yearly average minimum temperature value of `16^c`.
   #     - the same goes for the maximum temperature column. 
averages = Aus_weather[['Temp_Min', 'Temp_Max']].mean()
print(averages.round(1))
```

```python
# Short method to find mean for max 
# Min_df1 = Aus_weather.groupby('year')
# 
# Min_df1["Temp_Min", "Temp_Max"].mean()
```

```python
Min_df= Aus_weather.groupby(['Year'])['Temp_Min'].mean().reset_index().sort_values(by=['Year'], ascending=False)
Max_df= Aus_weather.groupby(['Year'])['Temp_Max'].mean().reset_index().sort_values(by=['Year'], ascending=False)
```

```python

Aus_weather['Temp_Min'] = Aus_weather.apply(lambda x : Min_df['Temp_Min'][Min_df['Year'] == x['Year']].mean() if np.isnan(x['Temp_Min']) else x['Temp_Min'], axis = 1)
print(Aus_weather['Temp_Min'].isnull().sum())
```

```python
Aus_weather['Temp_Max'] = Aus_weather.apply(lambda x : Max_df['Temp_Max'][Max_df['Year'] == x['Year']].mean() if np.isnan(x['Temp_Max']) else x['Temp_Max'], axis = 1)
print(Aus_weather['Temp_Max'].isnull().sum())
```

```python
# 8. Finally, add the degree centigrade symbol for both the minimum and the maximum temperates columns.
#  i.e 22.9 will be `22.9^c`
Aus_weather['Temp_Min'] = Aus_weather['Temp_Min'].map(str) + '°C' 
Aus_weather['Temp_Max'] = Aus_weather['Temp_Max'].map(str) + '°C' 
Aus_weather.head()
```
